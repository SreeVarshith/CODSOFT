```python
#UNDERSAMPLING
legit_sample = legit.sample(n=492)
```

```python
new_ds = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_ds['Class'] = new_ds['Class'].astype(int)
```

```python
new_ds.head()
```

```python
X = new_ds.drop(columns='Class',axis=1)
Y = new_ds['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction  = model.predict(X_train)
train_accuracy = accuracy_score(X_train_prediction,Y_train)
print('Accuracy for training data is ',train_accuracy *100)
```

```python
X_test_prediction  = model.predict(X_test)
test_accuracy = accuracy_score(X_test_prediction,Y_test)
print('Accuracy for testing data is ',test_accuracy *100)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Training Set Evaluation
print("Training Set:")
print("Confusion Matrix:\n", confusion_matrix(Y_train, X_train_prediction))
print("Classification Report:\n", classification_report(Y_train, X_train_prediction))
```

```python
# Testing Set Evaluation
print("\nTesting Set:")
print("Confusion Matrix:\n", confusion_matrix(Y_test, X_test_prediction))
print("Classification Report:\n", classification_report(Y_test, X_test_prediction))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Training Set Metrics
precision_train = precision_score(Y_train, X_train_prediction)
recall_train = recall_score(Y_train, X_train_prediction)
f1_train = f1_score(Y_train, X_train_prediction)

print("Training Set Metrics:")
print(f"Precision: {precision_train:.4f}")
print(f"Recall: {recall_train:.4f}")
print(f"F1-Score: {f1_train:.4f}")
```

```python
# Testing Set Metrics
precision_test = precision_score(Y_test, X_test_prediction)
recall_test = recall_score(Y_test, X_test_prediction)
f1_test = f1_score(Y_test, X_test_prediction)

print("\nTesting Set Metrics:")
print(f"Precision: {precision_test:.4f}")
print(f"Recall: {recall_test:.4f}")
print(f"F1-Score: {f1_test:.4f}")
```
